```python
text = clipboard.paste()
```

```python
# 제목, 타이틀, 해시태그 지정

title = re.search(r'제목: "(.*)"', text).group(1)
match = re.search(r'본문:(.*)', text, re.DOTALL)

if match:
    content, hashtag = match.group(1).split('#', 1)
    hashtag = '#'+hashtag
    content = content[:content.rfind('.')+1]
```
